# Importing the Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

# Loading the Dataset

```python
df = pd.read_csv( 'Diwali Sales Data.csv', encoding = 'unicode_escape')
```

#  Data Cleaning

```python
df.shape
```

```python
df.head(10)
```

```python
df.describe()
```

```python
df.info()
```

```python
df.drop(['Status', 'unnamed1'], axis = 1, inplace = True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtype
```

# Exploratory Data Analysis

# Gender

```python
df.columns
```

```python
ax = sns.countplot(x = 'Gender', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by= 'Amount', ascending = False)
sales_gen
```

```python
sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)
```

# Age


NOTE : FROM THE ABOVE GRAPH WE CAN SEE THAT MOST OF THE BUYERS ARE FEMALES AND EVEN THE PURCHASING POWER OF FEMALES ARE GREATER THAN MALES


```python
df.columns
```

```python
ax = sns.countplot(data = df , x= 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by= 'Amount', ascending = False)

sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age)
plt.show()
```


NOTE: FROM THE ABOVE GRAPH WE CAN SEE THAT MOST OF THE BUYERS ARE OF AGE GROUP BETWEEN 26-35 YEARS FEMALE


# State

```python
df.columns
```

```python
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by= 'Orders', ascending = False).head(10)

plt.figure (figsize = (15,5))
sns.barplot(x = 'State', y = 'Orders', data = sales_state)
plt.show()
```

```python
sales_state = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by= 'Amount', ascending = False).head(10)

plt.figure (figsize = (15,5))
sns.barplot(x = 'State', y = 'Amount', data = sales_state)
plt.show()
```


NOTE: FROM ABOVE GRAPHS WE CAN SEE THAT MOST OF THE ORDERS AND TOTAL SALES /AMOUNT ARE FROM UTTAR PRADESH, MAHARASHTRA AND KARNATAKA RESPECTIVELY


# Marital Status

```python
plt.figure (figsize = (7,5))
ax = sns.countplot(data = df, x = 'Marital_Status')
for bars in ax.containers:
    ax.bar_label(bars)
    
plt.show()    
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index = False)['Amount'].sum().sort_values(by= 'Amount', ascending = False).head(10)

plt.figure (figsize = (6,5))
sns.barplot(x = 'Marital_Status', y = 'Amount', hue = 'Gender', data = sales_state)
plt.show()
```


NOTE: FROM ABOVE GRAPH WE CAN SEE THAT MOST OF THE BUYERS ARE MARRIED(WOMEN) AND THEY HAVE HIGH PURCHASING POWER


# Occupation

```python
plt.figure (figsize = (20,5))
ax = sns.countplot(data = df, x = 'Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
    
plt.show()    
```

```python
sales_state = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by= 'Amount', ascending = False).head(10)

plt.figure (figsize = (15,5))
sns.barplot(x = 'Occupation', y = 'Amount', data = sales_state)
plt.show()
```


NOTE: FROM THE ABOVE GRAPH WE CAN SEE THAT MOST OF THE BUYERS ARE WORKING IN IT, HEALTHCARE AND AVIATION SECTOR


# Product Category

```python
plt.figure (figsize = (25,8))
ax = sns.countplot(data = df, x = 'Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
    
plt.show()    
```

```python
sales_state = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by= 'Amount', ascending = False).head(10)

plt.figure (figsize = (20,5))
sns.barplot(x = 'Product_Category', y = 'Amount', data = sales_state)
plt.show()
```


NOTE: FROM THE ABOVE GRAPH WE CAN SEE THAT MOST OF THE SOLD PRODUCTS ARE FROM FOOD, CLOTHING AND ELECTRONICS CATEGORY


# Product_Id

```python
sales_state = df.groupby(['Product_ID'], as_index = False)['Orders'].sum().sort_values(by= 'Orders', ascending = False).head(10)

plt.figure (figsize = (20,5))
sns.barplot(x = 'Product_ID', y = 'Orders', data = sales_state)
plt.show()
```
